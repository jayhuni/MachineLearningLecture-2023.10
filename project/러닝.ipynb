```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
df = pd.read_csv('data/팁309.csv')
df.head()
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.tail()
```

```python
model.encode(df.질문[0])
```

```python
df['embedding'] = pd.Series([[]] * len(df))
df.shape
```

```python
df.embedding = df.질문.map(lambda x: list(model.encode(x)))
df.head(3)
```

```python
df.to_csv('data/팁_dataset.csv', index=False)
```

- 테스트

```python
import json
wdf = pd.read_csv('data/팁_dataset.csv')
wdf.embedding = wdf.embedding.apply(json.loads)
wdf.head(3)
```

```python
text = '월세 알아볼때 어떻게 해요'
embedding = model.encode(text)
```

```python
wdf['유사도'] = wdf.embedding.map(lambda x: cosine_similarity([embedding],[x]).squeeze())
wdf.head(3)
```

```python
wdf.유사도.idxmax()
```

```python
wdf.iloc[wdf.유사도.idxmax()]
```

```python
wdf.답변[wdf.유사도.idxmax()]
```
