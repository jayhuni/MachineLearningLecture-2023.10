## 단순 선형회귀

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 아버지와 아들의 키 - Galton

```python
df = pd.read_csv('data/Galton.txt', sep='\t')
df.head()
```

- 전처리
    - 아버지와 아들의 키만 발췌
    - 컬럼명을 Father, Son 으로 변경
    - 단위(1 inch --> 2.54 cm) 변경

```python
# filtering
df = df[df.Gender == 'M']
df.head(3)
```

```python
# selection
df = df[['Father', 'Height']]
# 컬럼명 변경
df.columns = ['Father', 'Son']
df.tail(3)
```

```python
# 단위 변경
df = df * 2.54
df.tail(3)
```

- 산점도

```python
plt.figure(figsize=(5,5))
plt.scatter(df.Father, df.Son)
plt.xlabel("Father's height (cm)")
plt.ylabel("Son's height (cm)")
plt.title('Father vs Son', fontsize=15)
plt.grid()
plt.show()
```

- 회귀선

```python
weight, bias = np.polyfit(df.Father, df.Son, deg=1)     # X, Y, degree(차수)
weight, bias
```

```python
np.polyfit(df.Father, df.Son, 2)
```

```python
# X 축의 최소값, 최대값
df.Father.min(), df.Father.max()
```

```python
# 회귀선(직선)
xs = np.array([df.Father.min(), df.Father.max()])
ys = weight * xs + bias
ys
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df.Father, df.Son)
plt.plot(xs, ys, c='red', lw=3)
plt.xlabel("Father's height (cm)")
plt.ylabel("Son's height (cm)")
plt.title('Father vs Son', fontsize=15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.regplot(x='Father', y='Son', data=df)
plt.title('Father vs Son', fontsize=15)
plt.show()
```

- Scikit-Learn으로 회귀식 구하기

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.get_params()
```

```python
# 학습
lr.fit(df[['Father']], df.Son)      # X 값은 2차원이 되어야 함
```

```python
X = df.Father.values.reshape(-1, 1)
y = df.Son.values
lr.fit(X, y)
```

```python
# weight, bias
print(lr.coef_, lr.intercept_)
print(weight, bias)
```

- 평가 - R squared

```python
lr.score(X, y)
```

```python
from sklearn.metrics import r2_score
y_hat = lr.predict(X)
r2_score(y, y_hat)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_hat)
```
